# Get Data


```python
from mnist_data import MNIST

data = MNIST()
x_train = data.x_train
y_train = data.y_train
x_test = data.x_test
y_test = data.y_test
```

# Example of Image


```python
data.show_img(0)
```

# Check Shapes


```python
x_train.shape, y_train.shape
```

# Build Model


```python
from model import Neural_Network
import numpy as np

layers_dims = np.array([x_train.shape[0],3,4,1])

learning_rate = 0.075

model = Neural_Network(layers_dims)

```

# Train Model


```python
costs = model.train(x_train, y_train, learning_rate, num_iterations = 3000, print_cost=True)
```

# Plot Training


```python
from utilities import plot_costs, find_misclassified

plot_costs(costs, learning_rate)
```

# Check Accuracy on Test Set


```python
Y_pred_test = model.predict(x_test)
test_accuracy = model.accuracy(Y_pred_test, y_test)
print(f"Test Accuracy: {test_accuracy}%")
```

```python
# Assuming x_test, y_test are defined and model is an instance of Neural_Network
misclassified_indices = find_misclassified(x_test, y_test, Y_pred_test)
data.show_img(misclassified_indices[0])
```
